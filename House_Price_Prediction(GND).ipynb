```python
# the above resultt u can see there is a value which define with '-' in between number
# so what i can do just aplit them and calculate the mean of that number and fill them
# so for this - region it act as str or object so i try to change the datatype 

def convertRange(x):
    temp = x.split('-')
    if len(temp) == 2:
        return (float(temp[0]) + float(temp[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
def remove_outliers_sqft(df):
    df_output = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqFt)
        st = np.std(subdf.price_per_sqFt)
        gen_df = subdf[(subdf.price_per_sqFt > (m-st)) & (subdf.price_per_sqFt <= (m+st))]
        df_output = pd.concat([df_output, gen_df], ignore_index=True)
    return df_output
data = remove_outliers_sqft(data)
data.describe()
```

```python
def bhk_outlier_remove(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('size_BHK'):
            bhk_stats[bhk] = {
                'mean' : np.mean(bhk_df.price_per_sqFt),
                'std' : np.std(bhk_df.price_per_sqFt),
                'count' : bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('size_BHK'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqFt < (stats['mean'])].index.values)
    return df.drop(exclude_indices, axis = 'index')
```

```python
data = bhk_outlier_remove(data)
```

```python
# Separate Independant and dependant variable
X = data.drop(['price'], axis=1)
y = data['price']
```

```python
X.head()
```

```python
y.head()
```

```python
# Splitting the Dataset
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
# Linear Regression
column_trans = make_column_transformer((OneHotEncoder(sparse_output= False), ['location']), remainder = 'passthrough')
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,scaler, lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lr = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_lr)
```

### Lasso regression

```python
lasso = Lasso()
```

```python
pipe = make_pipeline(column_trans, scaler, lasso)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lasso = pipe.predict(X_test)
r2_score(y_test,y_pred_lasso)
```

### Ridge Regression

```python
ridge = Ridge()
```

```python
pipe = make_pipeline(column_trans, scaler, ridge)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred_ridge = pipe.predict(X_test)
r2_score(y_test,y_pred_ridge)
```

```python
print('No Regularization ', r2_score(y_test, y_pred_lr))
print('Redge: ', r2_score(y_test,y_pred_ridge))
print('Lasso: ', r2_score(y_test,y_pred_lasso))
```
